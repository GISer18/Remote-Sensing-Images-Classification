```python
import scipy
import h5py
from PIL import Image
from scipy import ndimage
import numpy as np
import random
from matplotlib import pyplot as plt
from scipy.misc import imread
from keras.preprocessing import image
from sklearn.model_selection import train_test_split

%matplotlib inline
```

```python
names_1 = ['airport','bareland','baseballfield','beach','bridge',
           'center','church','commercial','denseresidential','desert',
           'farmland','forest','industrial','meadow','mediumresidential']
names_2 = ['mountain','park','parking','playground','pond',
           'port','railwaystation','resort','river','school',
           'sparseresidential','square','stadium','storageTanks','viaduct']
nums_1 = [360,310,220,400,360,260,240,350,410,300,370,250,390,280,290]
nums_2 = [340,350,390,370,420,380,260,290,410,300,300,330,290,360,420]
```

```python
names = ['airport','bareland','baseballfield','beach','bridge',
         'center','church','commercial','denseresidential','desert',
         'farmland','forest','industrial','meadow','mediumresidential',
         'mountain','park','parking','playground','pond',
         'port','railwaystation','resort','river','school',
         'sparseresidential','square','stadium','storageTanks','viaduct']
```

```python
nums = [360,310,220,400,360,260,240,350,410,300,370,250,390,280,290,
        340,350,390,370,420,380,260,290,410,300,300,330,290,360,420]
indexs = [i for i in range(30)]
```

```python
random_index = random.sample(indexs, 15)
random_index = np.array(random_index)
random_index
```

```python
y1 = np.ones((nums[random_index[0]],1)) * random_index[0]
y2 = np.ones((nums[random_index[1]],1)) * random_index[1]
y3 = np.ones((nums[random_index[2]],1)) * random_index[2]
y4 = np.ones((nums[random_index[3]],1)) * random_index[3]
y5 = np.ones((nums[random_index[4]],1)) * random_index[4]
y6 = np.ones((nums[random_index[5]],1)) * random_index[5]
y7 = np.ones((nums[random_index[6]],1)) * random_index[6]
y8 = np.ones((nums[random_index[7]],1)) * random_index[7]
y9 = np.ones((nums[random_index[8]],1)) * random_index[8]
y10 = np.ones((nums[random_index[9]],1)) * random_index[9]
y11 = np.ones((nums[random_index[10]],1)) * random_index[10]
y12 = np.ones((nums[random_index[11]],1)) * random_index[11]
y13 = np.ones((nums[random_index[12]],1)) * random_index[12]
y14 = np.ones((nums[random_index[13]],1)) * random_index[13]
y15 = np.ones((nums[random_index[14]],1)) * random_index[14]
Y = np.vstack((y1,y2,y3,y4,y5,y6,y7,y8,y9,y10,y11,y12,y13,y14,y15))
Y = Y.astype('int64')
```

```python
Y.shape
```

```python
m = 0
print('id  nam  num')
for index in random_index:
    m = m + nums[index]
    print(index, names[index], nums[index])
print(m)
```

```python
X = []
path = 'AID/dataset/'
for index in random_index:
    sub_path = names[index] + '/'
    pic_name = names[index] + '_'
    for i in range(nums[index]):
        img = image.load_img(path + sub_path + pic_name + str(i+1) + ".jpg", target_size=(224, 224))
        img_arr = image.img_to_array(img)
        X.append(img_arr)
X = np.array(X)
```

```python
X.shape
```

```python
permutation = list(np.random.permutation(m))
X = X[permutation,:,:,:]
Y = Y[permutation,:]
```

```python
def convert_to_one_hot(Y, C):
    Y = np.eye(C)[Y.reshape(-1)].T
    return Y
```

```python
Y = convert_to_one_hot(Y, C=30).T
```

```python
Y.shape
```

```python
f = h5py.File('AID_15_classes.h5','w')
f['data'] = X
f['labels'] = Y
f.close()
```
